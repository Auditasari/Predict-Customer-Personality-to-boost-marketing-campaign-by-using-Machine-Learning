### Data Preprocessing 

```python
# agar tidak merubah file asli maka akan dilakukan copy dataframe
df1 = df.copy()
```

#### Handle Missing Value

```python
## Mengisi missing value dengan nilai median 
df1['Income'] = df1['Income'].fillna(df1['Income'].median())
```

#### Handle Data Type

```python
#### Mengganti tipe data kolom Year_Birth dan Dt_Customer
df1['Year_Birth'] = pd.to_datetime(df1['Year_Birth'], format='%Y')
df1['Dt_Customer'] = pd.to_datetime(df1['Dt_Customer'])
```

```python
df1['Year_Birth'] = df1['Year_Birth'].dt.year
```

### Feature Engineering

##### Total Transaction

```python
df1['Total_Transaction'] = df1[['NumDealsPurchases', 'NumWebPurchases','NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']].sum(axis=1)
df1['Total_Transaction'].describe()
```

```python
def transaction (Total_Transaction):
    if Total_Transaction < 8 :
        result = '0-7 Transaksi'
    elif Total_Transaction < 21 :
        result = '8-20 Transaksi'
    else : 
        result = '21-44 Transaksi'
    return result

df1['Transaction_Group'] = df1['Total_Transaction'].apply(transaction) 
```

##### Conversion Rate

```python

def conversion_rate (x,y):
  if y == 0:
    return 0
  return x/y

df1['Conversion_rate'] = round(df1.apply(lambda x : conversion_rate(x['Total_Transaction'], x['NumWebVisitsMonth']), axis =1),2) 
     
     
```

##### Mengelompokan Umur

```python
# mencari umur customer 
from datetime import datetime
# mengambil tahun pembelian customer
df1['Year'] = df1['Dt_Customer'].dt.year

#menghitung umur customer ketika melakukan pembelian
df1['Age'] = df1['Year'] - df1['Year_Birth']
df1['Age'].describe()
```

```python
# mengkelompokan umur 
def age_cat (Age):
    if Age < 18 :
        result = 'Teenager'
    elif Age < 64 :
        result = 'Adult'
    else : 
        result = 'Senior'
    return result

df1['Category_Age'] = df1['Age'].apply(age_cat) 
```

##### Mengelompokan Total Pembelian 

```python
df1['Total_Purchases'] = df1[['MntCoke', 'MntFruits', 'MntMeatProducts','MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)
def Spend (Total_Amount):
    if Total_Amount < 60000:
        result = 'Under 50 Ribu'
    elif Total_Amount < 260000:
        result = '60.000 - 250.000'
    elif Total_Amount < 600000:
        result = '260.000 - 500.000'
    elif Total_Amount < 1100000:
        result = '600.000 - 1.000.000'
    else : 
        result = 'Over 1.000.000'
    return result

df1['Spend_Group'] = df1['Total_Purchases'].apply(Spend)
```

#### Mengelompokan Jumlah Penerimaan Campaign

```python
df1['Total_Acc'] = df1[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']].sum(axis=1)
```

```python
def acc_campaign (Total_Acc):
    if Total_Acc >= 1 :
        result = 'Yes'
    else:
        result = 'No'
    return result
df1['Acc_Campaign'] = df1['Total_Acc'].apply(acc_campaign)
```

### Membuang Outlier 

```python
features = df1.select_dtypes(['float64','int64']).columns


for i in features:
    Q1 = df1[i].quantile(.25)
    Q3 = df1[i].quantile(.75)
    IQR = Q3 - Q1
    low = Q1 - (1.5 *IQR)
    high = Q3 + (1.5 * IQR)
    filter_data = ((df1[i] >= low) & (df1[i] <= high ))
    df1_clean = df1[filter_data]

df1_clean = df1_clean
print('Sebelum Dibuang Outlier', df1.shape)
print('Setelah Dibuang Outlier', df1_clean.shape)
```

```python
# Menduplikat data 
df2 = df1_clean.copy()
## menyatukan nilai janda dan duda ke menrikah 

df2['Marital_Status'] = df2['Marital_Status'].replace({'Janda': 'Menikah', 'Duda': 'Menikah'})

## Mengubah nilai Yes No pada kolom Acc_campaign menjadi 0 dan 1
df2['Acc_Campaign'] = df2['Acc_Campaign'].replace({'Yes': 1, 'No': 0})
```

```python
# Melakukan Label Encoder Education
df2.Education = df2.Education.map({
    'SMA' : 0,
    'D3': 1,
    'S1': 2,
    'S2' : 3,
    'S3' :4,
})

df2.Marital_Status = df2.Marital_Status.map({
    'Lajang' : 0,
    'Bertunangan': 1,
    'Menikah': 2,
    'Cerai' : 3

})

df2.Spend_Group = df2.Spend_Group.map({
    'Under 50 Ribu' : 0,
    '60.000 - 250.000': 1,
    '260.000 - 500.000': 2,
    '600.000 - 1.000.000' : 3,
    'Over 1.000.000' :4,
})

df2.Category_Age = df2.Category_Age.map({
    'Teenager' : 0,
    'Adult': 1,
    'Senior': 2
})

df2.Transaction_Group = df2.Transaction_Group.map({
    '0-7 Transaksi' : 0,
    '8-20 Transaksi': 1,
    '21-44 Transaksi': 2
})
```

### Melakukan Standardisasi

```python
df2 = df2.select_dtypes(exclude=['datetime64[ns]'])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = df2.drop('Acc_Campaign', axis = 1)
y = df2['Acc_Campaign']

X_std = scaler.fit_transform(X)
df_std = pd.DataFrame(data = X_std, columns= X.columns)
df_std.describe()
```

#### Modeling

#### Feature Selection

Untuk Menentukan Fitur apa yang akan digunakan, Saya menggunakan Metode RFM : <br>
* Recency = Recency
* Frequency = Total_Transaction
* Monetery = Total_Purchases

```python
df2_model = df_std[['Recency', 'Total_Transaction','Total_Purchases']]
df2_model
```

### Melakukan Clustering dan Melihat Jumlah Clustering Menggunakan Elbow Method

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df2_model.values)
    inertia.append(kmeans.inertia_)



plt.figure(figsize=(15, 8))
# plt.plot(inertia)

sns.lineplot(x=range(2, 11), y=inertia, color='seagreen', linewidth = 4)
sns.scatterplot(x=range(2, 11), y=inertia, s=300, color='gold',  linestyle='--')
plt.title('Jumlah Klaster Dengan Elbow Method', size = 14)
plt.xlabel('Jumlah Klater', size = 14)
```

```python
from sklearn.metrics import silhouette_score


range_n_clusters = list(range(2,11))
arr_silhouette_score_euclidean = []
for i in range_n_clusters:
    kmeans = KMeans(n_clusters=i).fit(df2_model)
    preds = kmeans.predict(df2_model)
     
    score_euclidean = silhouette_score(df2_model, preds, metric='euclidean')
    arr_silhouette_score_euclidean.append(np.round(score_euclidean,2))


fig, ax = plt.subplots(figsize=(8, 6))
sns.lineplot(x=range(2,11), y=arr_silhouette_score_euclidean, color='seagreen', linewidth = 3)
sns.scatterplot(x=range(2,11), y=arr_silhouette_score_euclidean, s=100, color='gold',  linestyle='--')
for i, label in enumerate(arr_silhouette_score_euclidean):
    plt.text(range(2,11) [i] + 0.2, arr_silhouette_score_euclidean[i], label, ha = 'center', va = 'bottom')
plt.title('Jumlah Klaster Berdasarkan Silhouette Score', size = 14)
plt.xlabel('Jumlah Klaster', size = 12)
plt.show()
```

Berdasarkan gambar elbow method dan gambar silhouette di atas, jumlah cluster yang optimal adalah 4, dimana titik siku berada di cluster 4 dan selisih nilai silhouette terkecil dan mendekati satu berada di cluster 4.

```python
# Melihat Jumlah Konsumen per Cluster
df_klaster.groupby(['Cluster'])['Km_lables'].count().reset_index(name = 'Total Konsumen').sort_values(by='Total Konsumen', ascending=False)
```
